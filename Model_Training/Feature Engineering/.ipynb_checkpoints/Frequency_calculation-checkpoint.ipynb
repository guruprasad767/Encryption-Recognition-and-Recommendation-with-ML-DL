```python
import pandas as pd
from collections import Counter
file_path = '../../Data/Processed/complete_dataset/encrypted_combined.csv'

# Load the CSV into a pandas DataFrame
df= pd.read_csv(file_path, quoting=1)

df
```

```python
df = df.rename(columns={'AES_encrypted': 'Encrypted Text'})
```

```python
df
```

```python
# Function to calculate % frequency of each character in a string
def char_frequency_percent(string):
    total_chars = len(string)
    char_counts = Counter(string)
    
    # Convert to percentage
    return {char: (count / total_chars) * 100 for char, count in char_counts.items()}

# Apply function and create a new DataFrame
freq_df = df['Encrypted Text'].apply(char_frequency_percent).apply(pd.Series).fillna(0)

# Merge with original DataFrame
df = pd.concat([df, freq_df], axis=1)

df.head(20)
```
