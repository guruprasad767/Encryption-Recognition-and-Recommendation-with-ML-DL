```python
import pandas as pd
from collections import Counter
file_path = '../../Data/Processed/complete_dataset/aes_blowfish_chacha.csv'

# Load the CSV into a pandas DataFrame
df= pd.read_csv(file_path, quoting=1)

df
```

```python
df['length'] = df['Encrypted'].apply(len)
```

```python
df
```

```python
unique_chars = set("".join(df['Encrypted']))

unique_chars
```

```python
for char in unique_chars:
    df[char] = df.apply(lambda row: (row['Encrypted'].count(char) / row['length']) * 100, axis=1)

df
```

```python
df.to_csv('vectorized(aes_chacha_.csv', index=False, quoting=1)
```
