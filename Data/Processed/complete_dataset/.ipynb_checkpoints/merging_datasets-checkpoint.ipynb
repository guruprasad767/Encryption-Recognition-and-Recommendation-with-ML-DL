```python
import pandas as pd

# Specify the path to your CSV file
file_path1 = '../algorithm_wise/aes_encrypted.csv'
file_path2 = '../algorithm_wise/chacha20_encrypted.csv'
file_path3 = '../algorithm_wise/blowfish_encrypted.csv'

# Load the CSV into a pandas DataFrame
df_aes = pd.read_csv(file_path1, quoting=1)
df_chacha20 = pd.read_csv(file_path2, quoting=1)
df_bf = pd.read_csv(file_path3, quoting=1)
```

```python
df_bf
```

```python
df_aes['algorithm'] = 'AES'
df_chacha20['algorithm'] = 'ChaCha20'
df_bf['algorithm'] = 'Blowfish'
```

```python
df_aes
```

```python
df_aes.rename(columns={'AES_encrypted': 'Encrypted'}, inplace=True)
df_chacha20.rename(columns={'ChaCha20_encrypted': 'Encrypted'}, inplace=True)
```

```python
# Concatenating all DataFrames
df_final = pd.concat([df_aes, df_chacha20, df_bf], axis=0, ignore_index=True)  

# Shuffling the DataFrame
df_final = df_final.sample(frac=1, random_state=37).reset_index(drop=True)

df_final.head(20)
```

```python
import csv

# Exporting to CSV with quoting
df_final.to_csv("aes_blowfish_chacha.csv", index=False, quoting=1)
```
